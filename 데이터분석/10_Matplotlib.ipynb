# Matplot을 이용한 시각화
- 파이썬으로 데이터를 시각화하는 데에는 matplotlib 라이브러리를 가장 많이 사용한다.
- matplotlib은 파이썬에서 2D 형태의 그래프, 이미지 등을 그릴 때 사용한다.
- 실제 과학 컴퓨팅 분야나 인공지능 분야에서도 많이 사용된다.

<br>

- matplotlib 모듈에는 다양한 모듈이 많이 있는데 그 중 가장 기본인 것이 pyplot 서브모듈이다

```python
import matplotlib.pyplot as plt
```

### 1. plot() 
- 직선 혹은 꺽은선 그래프

```python
np.random.seed(100)
plt.figure(figsize=(12,4))
plt.grid()
plt.axis([0,100,0,1000]) #axis([xmin, xmax, ymin, ymax])
plt.plot(range(1,101), np.random.randint(1,1001,100), 'k-')
```

### 1_1. 겹쳐진 그래프

```python
t = np.arange(0, 5, 0.2)

plt.plot(t,t,'r--')
plt.show()
```

```python
plt.plot(t,t,'r--', t,t**2,'bo-',t,t**3,'k', t,t**4,'sg-', markersize=4)
plt.show()
```

```python
plt.plot(t,t,'r--', label='a')
plt.plot(t,t**2,'bo-',label='b')
plt.plot(t,t**3,'k',label='c')
plt.plot(t,t**4,'sg-', markersize=4, label='d')
plt.legend()
plt.show()
```

```python
t = np.arange(0,12, 0.01)
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), 'r', linewidth=6, label='Sin')
plt.plot(t, np.cos(t), 'b', label='Cos')

plt.legend()
plt.grid()

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.xlabel('time')
plt.ylabel('Amplitude')

plt.title('Graph test')
plt.show()
```

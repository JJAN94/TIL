```python
def load_csv(year) :
    dir_name = '../data/d/서울특별시_부동산_실거래가_정보_{}년.csv'.format(year)
    df_name = pd.read_csv(dir_name, encoding='cp949')
    df_name.drop(columns=['실거래가아이디', '지번코드', '물건번호', '업무구분', '업무구분코드'], inplace=True)
    df_name['물건금액(만원)'] = df_name['물건금액']/10000
    df_name['해당년'] = year
    return df_name
```

```python
df_2006 = load_csv(2006)
df_2007 = load_csv(2007)
df_2008 = load_csv(2008)
df_2009 = load_csv(2009)
df_2010 = load_csv(2010)
df_2011 = load_csv(2011)
df_2012 = load_csv(2012)
df_2013 = load_csv(2013)
df_2014 = load_csv(2014)
df_2015 = load_csv(2015)
df_2016 = load_csv(2016)
df_2017 = load_csv(2017)
df_2018 = load_csv(2018)
df_2019 = load_csv(2019)
df_2020 = load_csv(2020)
```

```python
def group_price(df, year) :
    df = df.pivot_table(index=['자치구명', '법정동명', '건물주용도'], values='물건금액(만원)')
    df.rename(columns={'물건금액(만원)' : '{}년 물건가격(만원)'.format(year)}, inplace=True)
    return df
```

```python
price_2006 = group_price(df_2006, 2006)
price_2007 = group_price(df_2007, 2007)
price_2008 = group_price(df_2008, 2008)
price_2009 = group_price(df_2009, 2009)
price_2010 = group_price(df_2010, 2010)
price_2011 = group_price(df_2011, 2011)
price_2012 = group_price(df_2012, 2012)
price_2013 = group_price(df_2013, 2013)
price_2014 = group_price(df_2014, 2014)
price_2015 = group_price(df_2015, 2015)
price_2016 = group_price(df_2016, 2016)
price_2017 = group_price(df_2017, 2017)
price_2018 = group_price(df_2018, 2018)
price_2019 = group_price(df_2019, 2019)
price_2020 = group_price(df_2020, 2020)
```

```python
price_df = pd.concat([price_2006, price_2007, price_2008, price_2009, price_2010, price_2011, price_2012, price_2013, price_2014, price_2015, price_2016, price_2017, price_2018, price_2019, price_2020 ], axis=1)
dfdf = price_df.T
dfdf.head()
```

```python
plt.figure(figsize=(12,6))
a = price_df.loc['도봉구'].loc['창동'].T
plt.plot(a.index, a['단독주택'],'g')
plt.plot(a.index, a['연립주택'],'r')
plt.plot(a.index, a['오피스텔'],'b')
plt.plot(a.index, a['아파트'], 'k')
plt.xticks(rotation=45)
plt.legend(['단독주택', '연립주택', '오피스텔', '아파트'])
plt.grid()
plt.show()
```

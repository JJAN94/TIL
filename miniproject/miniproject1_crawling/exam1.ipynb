```python
def get_url() :
    binary = 'c:\\chromedriver\\chromedriver.exe'
    # 브라우저를 인스턴스화 하자.
    
    # 원하는 년, 월, 일 선택
    today = str(datetime.datetime.now())[:10].split('-')
    print('뽑고 싶은 연도를 선택하세요. (1990/01/01 ~ {}/{}/{})'.format(today[0], today[1], today[2]))
    startdate = input('언제부터 검색할까요? : ')
    enddate = input('어디까지 검색할까요? : ')
    start_date=startdate.split('/')
    end_date=enddate.split('/')
    print(type(start_date[2]))
    # 검색어
    s_name = input('원하는 검색어는 ? :')
    browser = webdriver.Chrome(binary)
    browser.get('https://www.naver.com/')
    
    mouse1 = browser.find_element_by_class_name('input_text')
    mouse1.send_keys(s_name)
    mouse1.submit()

    browser.find_element_by_link_text('이미지').click()
    time.sleep(5)
    browser.find_element_by_xpath('//*[@id="snb"]/div[1]/div/div[3]/a').click()
    time.sleep(3)
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[1]/a[8]').click()
    time.sleep(2)
    # 시작년도 클릭
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[1]/span[1]/a').click()
    time.sleep(1)
    # 년도 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[1]/div/div/div/ul/li[{}]/a'.format(int(start_date[0])-1989)).click()
    time.sleep(2)
    # 월 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[{}]/a'.format(start_date[1])).click()
    time.sleep(3)
    # 일 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[{}]/a'.format(start_date[2])).click()
    time.sleep(2)
    # 끝 년도 클릭
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[1]/span[3]/a').click()
    time.sleep(2)
    # 년도 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[1]/div/div/div/ul/li[{}]/a'.format(int(end_date[0])-1989)).click()
    time.sleep(2)
    # 월 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[2]/div/div/div/ul/li[{}]/a'.format(end_date[1])).click()
    time.sleep(2)
    # 일 선택
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[2]/div[3]/div/div/div/ul/li[{}]/a'.format(end_date[2])).click()
    time.sleep(2)
    
    # 기간 선택 완료
    browser.find_element_by_xpath('//*[@id="snb"]/div[2]/ul/li[2]/div/div[2]/div[3]/button').click()
    time.sleep(2)
    
    for i in range(1,2) : # 5번 스크롤 내려가는 상태
        browser.find_element_by_xpath("//body").send_keys(Keys.END)
        time.sleep(3) # END 키 누르고 내려가는데 시간이 걸려서 sleep 해줌. 키보드에서 END키를 총 5번 누르니까 End1번 누르고 10초 쉰다.

    time.sleep(5) # 네트워크 느릴까봐 안정성을 위해 sleep을 해줌. # 이거 안하면 이미지 안뜰 수 도 있음.
    source = browser.page_source # 크롬창 닫히기 전에 소스코드를 리턴받는다.
    soup = BeautifulSoup(source, "lxml")

    return soup

if __name__ == '__main__' :
    ls = get_url()
    print(ls)
```

```python
def html_img() :
    params = []
    soup = get_url()
    imgList = soup.find_all('img', class_='_image _listImage')
    for i in imgList :
        try :
            params.append(i['src'])
        except :
            params.append(i['data-src'])
    return params
```

```python
def fetch_detail_url() :
    params = html_img()
    
    for idx, p in enumerate(params, 1) :
        urllib.request.urlretrieve(p, 'C:\\Users\\JAY\\Desktop\\TIL\\miniproject\\miniproject1_crawling\\download_folder\\'+str(idx)+'.jpg')
```

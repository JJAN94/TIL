# Opencv 기초 사용법

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

- Opencv는 영상 데이터를 numpy.ndarray로 표현한다.

```python
# img1, img2는 numpy 배열로 되어있다.
img1 = cv2.imread('./cv/ch01/cat.bmp', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./cv/ch01/cat.bmp', cv2.IMREAD_COLOR)
```

- numpy.ndarray
    - ndim : 차원 수, len(img.shape)와 같음
    - shape : 각 차원의 크기 (h,w)(grayscale) 또는 (h,w,3)(color)
    - size : 전체 원소 개수
    - dtype : 원소의 데이터 타입, 영상 데이터는 unit8

- 영상 속성과 픽셀 값
    - 그레이스케일 영상 : cv2.CV_8UC1 -> numpy.uint8, shape=(h,w)
    - 컬러 영상 : cv2.CV_8UC3 -> numpy.uint8, shape=(h,w,3)

    - opencv 영상 데이터 자료형과 numpy 자료형
    | Opencv 자료형 (1채널) | Numpy 자료형 | 구분 |
    | :-- | :-- | :-- |
    | cv2.CV_8U | numpy.uint8 | 8비트 부호없는 정수 |
    | cv2.CV_8S | numpy.int8 | 8비트 부호있는 정수 |
    | cv2.CV_16U | numpy.uint16 | 16비트 부호없는 정수 |
    | cv2.CV_16S | numpy.int16 | 16비트 부호있는 정수 |
    | cv2.CV_32S | numpy.int32 | 32비트 부호있는 정수 |
    | cv2.CV_32F | numpy.float32 | 32비트 부동소수형 |
    | cv2.CV_64F | numpy.float64 | 64비트 부동소수형 |
    | cv2.CV_16F | numpy.float16 | 16비트 부동소수형 | 

```python
img1 = cv2.imread('./cv/ch01/cat.bmp', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('./cv/ch01/cat.bmp', cv2.IMREAD_COLOR)
print('type(img1):', type(img1)) # type(img1): <class 'numpy.ndarray'>
print('img1.shape:', img1.shape) # img1.shape: (480, 640)
print('img2.shape:', img2.shape) # img2.shape: (480, 640, 3)
print('img2.dtype:', img2.dtype) # img2.dtype: uint8

if len(img1.shape) == 2:
    print('img1 is a grayscale image')
elif len(img1.shape) == 3:
    print('img1 is a truecolor image')

if img2.ndim == 2:
    print('img2 is a grayscale image')
elif img2.ndim == 3:
    print('img2 is a truecolor image')

    
# 출력되는 값은 height, width로 나온다.(세로, 가로) (행, 열)
h1,w1 = img1.shape[:2]
print('img1 size: {} x {}'.format(w1, h1))

h2,w2 = img2.shape[:2] # 컬러 영상인 경우 채널도 있으서 슬라이싱이 없으면 에러가 난다.
print('img2 size: {} x {}'.format(w2, h2))
```

```python
# 픽셀값 참조
x = 20
y = 10
p = img1[y, x]
print(p) # 해당 좌표에 해당하는 픽셀 값

# for y in range(h1):
#     for x in range(w1):
#         img1[y,x] = 255
#         img2[y,x] = (0, 255, 255)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()

# for문으로 픽셀 값을 변경하는 작업은 매우 느리므로, 픽셀 값 참조 방법만 확인하고 실제로는 사용 금지
```

```python
img1[:,:] = 0
img2[:,:] = (0, 255, 255)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(img1)
plt.subplot(1,2,2)
plt.imshow(img2)
plt.show()
```

# 키워드(keyword)
- 특별한 의미가 부여된 단어, 파이썬에서 미리 쓰겠다고 한 단어들

- help() -> "keywords" (Lib폴더 내)
- import 모듈명 (C:\Users\JAY\AppData\Local\Programs\Python\Python38\Lib\keywords.py)
    - import keyword 하고 dir()하면 목록에 keyword가 생김. (dir : 목록 확인)
    - dir(keyword) 목록 확인
    - keyword.kwlist 하면 keyword가 가진 목록 리스트를 띄어준다.
    - del keyword (메모리에서 해당 객체를 삭제함, 삭제 후 dir()하면 목록에서 사라짐)
    - del 객체 : 객체를 메모리에서 삭제

# 식별자(identifier)
- 변수(객체 변수), 함수, 메서드, 클래스, 모듈 등을 구분하기 위해 붙이는 이름을 식별자라고 한다.
- 식별자 특징
    - 어떤 대상을 유일하게 구별하는 이름
    - 파이썬의 식별자는 유니코드(unicode) 형식을 취한다.
    - 코드에 존재하는 변수, 자료형, 서브루틴등을 가르키는 토큰
    - 프로그램 정보를 처리하기 위해서 그 정보를 가르키는 방법으로 사용된다.

- 식별자 규칙
    - 영문자 소, 대문자, 숫자, _, 를 사용할 수 있다. (대소문자는 구분된다.)
    - 단, 숫자로 시작하지는 못한다.
    - @, \$, % 등 문자 부호는 사용하지 않는다.
    - 클래스명은 시작하는 대문자를 제외하고 모두 소문자이다.
    - _ : private
    - __ : strong private
    - \__로 끝나는 식별자 언어에 의해 정의된 특별한 이름을 의미한다. ex) '\__name__'
    - 키워드를 제외하고 이름을 명명한다.
- 컨벤션
    - snake case : 변수, 함수 이름에 사용 ex) play_data
    - camel case : 클래스 이름에 사용 ex) PlayDate, Playdata

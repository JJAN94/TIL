## 4. 텍스트 & 엔트리

```python
root = Tk()

# 창 title 지정
root.title('프로젝트를 하자')
# 창 크기 조절
root.geometry('810x540')

txt = Text(root, width=30, height=5) # 텍스트 위젯 생성
txt.pack()

txt.insert(END, '글자를 입력하세요 : ') # 텍스트 위젯에 넣을 수 있는 글자 맨 처음에 지정
# txt는 여러줄 입력받을 때 쓴다고 생각하자.

e = Entry(root, width=30) # 한줄입력받을때.
e.pack()
e.insert(0, '한 줄만 입력하시오.')


def btncmd() :
    
    # 내용출력
    print(txt.get('1.0', END)) # 1.0 = 1번째줄 0번째 column위치 부터 ~ END까지 get하자 
    print(e.get())
    
    # 내용삭제
    txt.delete('1.0', END)
    e.delete(0, END)
    
btn = Button(root, text='클릭', command=btncmd)
btn.pack()


root.mainloop()
```

## 8. 콤보박스

```python
# 콤보박스는 새로 임포트를 해줘야한다.
import tkinter.ttk as ttk

root = Tk()

# 창 title 지정
root.title('프로젝트를 하자')
# 창 크기 조절
root.geometry('810x540')



values = [str(i) + '일' for i in range(1, 32)]
combobox = ttk.Combobox(root, height=5, values=values)
combobox.pack()
combobox.set('카드 결제일') # 최초 목록 제목 설정 뿐 아니라 버튼 클릭을 통한 값 설정도 가능.. 오오...


# state에 제약조건을 주면 선택지외에 다른 값은 입력하지 못한다.
# height는 눌렀을 때 몇개까지 목록으로 보여줄 것인지를 나타낸다.
readonly_combobox = ttk.Combobox(root, height=20, values=values, state='readonly')
readonly_combobox.current(0) # 0번째 인덱스 값 선택
readonly_combobox.pack()

def btncmd() :
    print(combobox.get()) # 선택된 값 표시
    print(readonly_combobox.get())
   
    
btn = Button(root, text='선택', command=btncmd)
btn.pack()


root.mainloop()
```

```python
# 프로그레스 바는 새로 임포트를 해줘야한다.
import time
import tkinter.ttk as ttk

root = Tk()

# 창 title 지정
root.title('프로젝트를 하자')
# 창 크기 조절
root.geometry('810x540')

p_var2 = DoubleVar() # 실수형
progressbar2 = ttk.Progressbar(root, maximum=100, length=150, variable=p_var2)
progressbar2.pack()

def btncmd2() :
    for i in range(1,101) :
        time.sleep(0.01) # 0.01초 딜레이
        
        p_var2.set(i) #프로그레스바의 값 설정
        
#########################################################################################################

        progressbar2.update() # 매번 실행 할 때마 gui에 반영한다. (UI 업데이트)
        if i == 99 :
            time.sleep(1)
            print(p_var2.get(),'퍼센트 완료... 거의 완료되었습니다.')
            
    
#########################################################################################################        
   
    
btn = Button(root, text='시작', command=btncmd2)
btn.pack()

root.mainloop()
```

## 11. 메시지 박스

```python
# 메시지 박스는 임포트를 해줘야한다.
import tkinter.messagebox as msgbox

root = Tk()

# 창 title 지정
root.title('프로젝트를 하자')
# 창 크기 조절
root.geometry('810x540')


#기차 예메 시스템이라고 가정
def info() :
    msgbox.showinfo('알림', '정상적으로 예매 완료되었습니다.')
    
def warn() :
    msgbox.showwarning('경고', '해당 좌석은 매진되엇습니다.')

def error() :
    msgbox.showerror('에러', '결제 오류가 발생했습니다.')
    
def okcancel() :
    msgbox.askokcancel('확인 취소', '입석만 가능합니다. 티켓팅 하시겠습니까?') # 선택지는 자동으로 주는건가보다
    
def retrycancel() :
    msgbox.askretrycancel('재시도 / 취소', '일시적인 오류입니다. 다시 시도하기겠습니까?')

def yesno() :
    msgbox.askyesno('예 / 아니오', '자고 싶니?') # 예...자고싶어요...
    
def yesnocancel() :
    response = msgbox.askyesnocancel(title=None, message='졸립다..') # 타이틀에 None 주면 창 이름이 없음
    # 네 : 저장 후 종료
    # 아니오 : 저장 하지 않고 종료
    # 취소 : 프로그램 종료 취소(현재 화면에서 계속 작업)
    print('응답 : ', response) # True, False, None -> 예=1, 아니오=0, 그 외
    if response == 1 :
        print('예')
    elif response == 0 :
        print('아니오')
    else :
        print('취소')
        
#         버튼에 대한 값을 받고 준다.
    
    
    
    
Button(root, command=info, text='알림').pack()
Button(root, command=warn, text='경고').pack()
Button(root, command=error, text='에러').pack()
Button(root, command=okcancel, text='확인 취소').pack()
Button(root, command=retrycancel, text='재시도 취소').pack()
Button(root, command=yesno, text='예 아니오').pack()
Button(root, command=yesnocancel, text='예 아니오 취소').pack()

root.mainloop()
```

#### 그리드 심화
- grid 옵션에 sticky가 있음
- sticky=N+E+W+S
    - 그리드끼리 동서남북으로 방향지정 후 달라붙게 해서 딱 그리드를 채울 수 있게 해주는 기능

- padx, pady : 단, 글자가 길면 글자에 맞춰 늘어나기에 버튼간 크기가 조금씩 다를 수도 있다.
    - Button 옵션에 padx, pady를 줘서 버튼 크기를 키워주자.

## 15. 퀴즈 - 메모장 만들기

```python
import os # 파일 있는지 없는지 체크하기 위해서
from tkinter import *

root = Tk()
root.title('제목없음 - Windows 메모장')
root.geometry('640x480')


# 열기, 저장 파일 이름
filename = 'mynote.txt'

def f_open() :
    if os.path.isfile(filename) : # 파일 있으면 true, 없으면 false
        with open(filename, 'r', encoding='utf8') as file :
            txt.delete('1.0', END) # 텍스트 위젯 본문 삭제
            txt.insert(END, file.read()) # 삭제후 파일 내용을 본문에 입력
def f_save() :
    with open(filename, 'w', encoding='utf8') as file :
        file.write(txt.get('1.0', END)) # 모든 내용을 가져와서 저장


# 파일 메뉴 만들기
menu = Menu(root)

menu_file = Menu(menu, tearoff=0)
menu_file.add_command(label='열기', command=f_open)
menu_file.add_command(label='저장', command=f_save)
menu_file.add_separator()
menu_file.add_command(label='끝내기', command=root.quit)
menu.add_cascade(label='파일', menu=menu_file)

# 편집, 서식, 보기, 도움말
menu.add_cascade(label='파일')
menu.add_cascade(label='서식')
menu.add_cascade(label='보기')
menu.add_cascade(label='도움말')

# 스크롤 바
scrollbar = Scrollbar(root)
scrollbar.pack(side='right', fill='y')


# 본문 영역
txt = Text(root, yscrollcommand=scrollbar.set)
txt.pack(side='left', fill='both', expand=True)

scrollbar.config(command=txt.yview)



root.config(menu=menu)
root.mainloop()
```

```python
import matplotlib.pyplot as plt
```

```python
o1, o2 = 0, 0
x, y = 10, 15
a1, b1 = 3/5, 4/5
a2, b2 = -4/5, 3/5
```

```python
xp = x*a1 + y*b1
yp = x*a2 + y*b2
```

```python
plt.figure(figsize=(10,5))
plt.plot([o1,x], [o2,y], 'k')
plt.plot([o1,a1], [o2,b1], 'r')
plt.plot([o1,a2], [o2,b2], 'r')
plt.plot([o1,xp], [o2,yp], 'b')
```

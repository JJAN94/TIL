# PySpark Documentation

PySpark is an interface for Apache Spark in Python. It not only allows you write Spark applications using Python APIs, but also provides the PySpark shell for interactively analyzing your data in a distributed environment. PySpark supports most of Spark's features such as Spark SQL, DataFrame, Streaming, MLlib ans Spark Core.

PySpark는 파이썬의 아파치 스파크를 위한 인터페이스이다. Python API를 사용하여 Spark 애플리케이션을 작성할 수 있을 뿐만 아니라 분산 환경에서 데이터를 대화식으로 분석할 수 있는 PySpark 셸을 제공합니다. PySpark는 스파크 SQL, 데이터프레임, 스트리밍, MLlib 및 스파크 코어 등 스파크의 기능 대부분을 지원한다.

## Spark SQL and DataFrame
Spark SQL is a Spark module for structured data processing. It provides a programming abstraction called DataFrame and can also act as distributed SQL query engine.

Spark SQL은 구조화된 데이터 처리를 위한 Spark 모듈입니다. DataFrame이라는 프로그래밍 추상화를 제공하며 분산 SQL 쿼리 엔진 역할을 할 수도 있다.

## pandas API on Spark
pandas API on Spark allows you to scale your pandas workload out. With this package, you can :
- Be immediately productive with Spark, with no learngin curve, if you are already familiar with pandas.
- Have a single codebase that works both with pandas (tests, smaller datasets) and with Spark (distributed datasets).
- Switch to pandas API and PySpark API contexts easily without any overhead.

판다 API on Spark를 사용하면 판다의 작업량을 확장할 수 있습니다. 이 패키지로 다음을 수행할 수 있습니다.
- 판다스에 대해 이미 잘 알고 있다면, 학습 곡선 없이 스파크로 즉시 생산성을 높이십시오
- 판다스(테스트, 소규모 데이터 세트) 및 스파크(분산 데이터 세트)에서 모두 작동하는 단일 코드베이스가 있어야 합니다.
- 오버헤드 없이 판다스 API와 PySpark API 컨텍스트로 쉽게 전환할 수 있습니다.

## Streaming
Running on top of Spark, the streaming feature in Apache Spark enables powerful interactive and analytical applications across both streaming and historical data, while inheriting Spark’s ease of use and fault tolerance characteristics.

스파크 위에서 실행되는 Apache Spark의 스트리밍 기능은 스파크의 사용 편의성과 내결함성 특성을 계승하면서 스트리밍 및 과거 데이터 모두에서 강력한 인터랙티브 및 분석 애플리케이션을 가능하게 합니다.

## MLlib
Built on top of Spark, MLlib is a scalable machine learning library that provides a uniform set of high-level APIs that help users create and tune practical machine learning pipelines.

스파크 위에 구축된 MLlib는 사용자가 실용적인 머신러닝 파이프라인을 만들고 튜닝할 수 있도록 도와주는 통일된 고급 API 세트를 제공하는 확장 가능한 머신러닝 라이브러리이다.

## Spark Core
Spark Core is the underlying general execution engine for the Spark platform that all other functionality is built on top of. It provides an RDD (Resilient Distributed Dataset) and in-memory computing capabilities.

스파크 코어는 다른 모든 기능들이 위에 구축되는 스파크 플랫폼의 기본 일반 실행 엔진이다. RDD(Resolient Distributed Dataset) 및 인메모리 컴퓨팅 기능을 제공합니다.

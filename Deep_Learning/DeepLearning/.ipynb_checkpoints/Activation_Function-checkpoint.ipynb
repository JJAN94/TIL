```python
import matplotlib.pyplot as plt
import numpy as np
```

# Activation Function (활성함수)

## Sigmoid Function (시그모이드)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, num=100)
```

```python
y = sigmoid(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.grid()
plt.show()
```

## ReLU Function (렐루 - Rectified Linear Unit)

```python
def ReLU(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-5, 5, 100)
y = ReLU(x)
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.grid()
plt.show()
```

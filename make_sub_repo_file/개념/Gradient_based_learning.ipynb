```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
func_x = np.linspace(-3,3,100)
func_y = func_x**2
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(func_x, func_y)

x = 2
lr = 0.1

for _ in range(10):
    dy_dx = 2*x
    x = x - (lr*dy_dx)
    y = x**2
    ax.scatter(x, y, color='red', s=100)
```

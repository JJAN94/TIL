# 영상의 산술 연산


````python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
````

## 덧셈연산

### 사용법
- 기본적으로 add 함수를 써서 saturate 연산을 해준다.
- 기본적으로 add 함수의 src1, src2의 타입이 같아야 한다. 다를 경우 dtype(cv2.타입)인자에 타입을 주어야한다.

### 가중치 합 (weighted sum)
$ dst(x,y) = saturate(\alpha\cdot src1(x,y) + \beta\cdot src2(x,y))$
- 그냥 더할 경우 255를 넘어가는 경우가 많아서 이를 보정해 주기 위해서 사용
- $\alpha + \beta = 1$이 되도록 설정 => 두 입력 영상의 평균 밝기를 유지

### 평균 연산(average)
- 가중치를 $\alpha = \beta = 0.5$로 설정한 가중치 합
$ dst(x,y) = \frac{1}{2}(src1(x,y) + src2(x,y))$

#### 사용법
```python
cv2.addWeighted(src1, alpha, src2, beta, gamma, dst=None, dtype=None)
```

- src1, src2 : 1,2번째 영상
- alpha, beta : 1,2번째 가중치
- gamma : 결과 영상에 추가적으로 더할 값

## 뺄셈 연산
$ dst(x,y) = saturate(src1(x,y) - src2(x,y)) $
- src1과 src2의 연산 순서에 따라 결과가 달라지니 유의
- 결과가 0보다 작으면 다 0으로 설정

### 사용법
```python
cv2.subtract(src1, src2, dst=None, mask=None, dtype=None)
```
- src1, src2 : 1, 2번째 영상 또는 스칼라

## 차이연산
$ dst(x,y) = \left\vert src1(x,y) - src2(x,y) \right\vert $
- 두 영상의 같은 위치에 존재하는 픽셀 값에 대하여 뺄셈 연산을 수행한 후, 그 절대값을 결과 영상의 픽셀 값으로 설정
- 뺄셈 연산과 다르게 계산 순서에 영향 X

### 사용법
```python
cv2.absdiff(src1, src2, dst=None)
```

````python
src1 = cv2.imread('../src/ch03/lenna256.bmp', cv2.IMREAD_GRAYSCALE)
src2 = cv2.imread('../src/ch03/square.bmp', cv2.IMREAD_GRAYSCALE)

if (src1 is None) or (src2 is None):
    print('Image load failed')
    sys.exit()
    
dst1 = cv2.add(src1, src2, dtype=cv2.CV_8U)
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0)
dst3 = cv2.subtract(src1, src2)
dst4 = cv2.absdiff(src1, src2)
plt.figure(figsize=(10,6))
plt.subplot(2,3,1), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(2,3,2), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')
plt.subplot(2,3,3), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('add')
plt.subplot(2,3,4), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('addWeighted')
plt.subplot(2,3,5), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('subtract')
plt.subplot(2,3,6), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('absdiff')
plt.show()
````

```python
# 3-4 예제 데이터의 성별을 1과 0으로 라벨링 처리 하기
df['성별'].value_counts() # 합계
```

```python
df['성별'].value_counts(normalize=True) # 비율로 출력
```

```python
df['성별(라벨)']= df['성별'].replace('여', 0).replace('남', 1)
df[['성별', '성별(라벨)']]
```

```python
# 3-5 구매 수량이 음수(-)인 값들만 따로 추출해 새로운 테이블로 선언
new_df = df[df['구매수량'] < 0]
print(new_df.shape)
new_df.head()
```

```python
# 3-6 예제 데이터에서, '회원 번호'를 기준으로 '구매 금액'의 합을 계산하고, 가장 많이 구매를 한 고객 5명을 확인하기
df.pivot_table(index='회원번호', values='구매금액', aggfunc=sum).sort_values(by='구매금액', ascending=False).head()
```

```python
# 3-7 예제 데이터에서 '요일'별로 '물품대분류'의 '구매금액'의 합을 계산하고, 판매량이 가장 높은 요일과, 해당 요일에 가장 많이 판매된 품목을 확인하기
df.pivot_table(index='요일', values='구매금액', aggfunc=sum).idxmax()
```

```python
df.pivot_table(index=['요일', '물품대분류'], values='구매금액', aggfunc=sum).loc['월'].idxmax()
```

```python
# 3-8 예제 데이터에서 '연령대' 별로 '구매금액'의 평균을 계산하고, 구매금액의 평균이 가장 높은 연령대 확인하기
df.pivot_table(index='연령대', values='구매금액').sort_values(by='구매금액').tail(1)
```

```python
df.pivot_table(index='연령대', values='구매금액').iloc[np.argmax(df.pivot_table(index='연령대', values='구매금액'))]
```

```python
df.pivot_table(index='연령대', values='구매금액').idxmax()
```

```python
# 3-9 예제 데이터에서 일별로 '구매금액'의 총합을 계산하고, 가장 적게 팔린 날을 확인하기
sub_df = df.pivot_table(index='요일', values='구매금액', aggfunc=sum)
sub_df
```

```python
sub_df.idxmin()
```

```python
# 3-10 예제 데이터에서 '공급주차'별로 '구매금액'의 총합을 계산하고 가장 많이 팔린 '공급주차'를 확인하기
df.pivot_table(index='공급주차', values='구매금액', aggfunc=sum).sort_values(by='구매금액', ascending=False)
```

```python
df.pivot_table(index='공급주차', values='구매금액', aggfunc=sum).idxmax()
```

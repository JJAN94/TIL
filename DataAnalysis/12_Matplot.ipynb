```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# 히스토그램
- plt.hist()
- 데이터의 분포 상태를 나타내는 그래프로써 데이터의 빈도에 따라서 막대의 높이가 결정된다.

```python
plt.hist([1,1,2,3,4,5,6,6,7,8,10], bins=30)
plt.show()
```

**bins**
- 해당 막대의 영역(bin)를 얼마나 채우는지 결정해주는 변수이다.
- 정수가 들어가면 해당 정수+1을 범위로 나눈 값으로 너비를 입력한다

```python
# bins 옵션
dice = []
for i in range(10) :
    dice.append(np.random.randint(1,7))
# print(dice)
plt.hist(dice)
# sns.kdeplot(dice)
plt.show()
```

```python
# 큰수의 법칙
# 던지는 횟수를 늘릴수록 확률이 1/6으로 수렴함

dice = []
for i in range(999999) :
    dice.append(np.random.randint(1,7))
plt.hist(dice)
plt.show()
```

```python
mu, sigma = 100, 15

x = mu+sigma*np.random.randn(99999)
# plt.hist(x)
sns.kdeplot(x)
# plt.show()
```

```python
mu, sigma = 100, 15

x = mu+sigma*np.random.randn(999999)
n, bins, patches = plt.hist(x, bins=200, facecolor='g', density=1,alpha=0.1)
# alpha : 투명도
# density : 1을 기준으로 했을 때 상대적 빈도값
plt.grid()
plt.text(40,0.025, r'$\mu = 100, \sigma = 15$')
plt.show()
```

# Boxplot

```python
# 0~100까지 총 101개의 숫자를 만든다.
xs = np.array(np.linspace(0,100, 101))

df = pd.DataFrame(xs, columns=['feature'])
```

```python
# 위에서 만든 df를 이용해서 boxplot을 그려보자.

# 출력 그래프 보면, 25, 75% 사이에 박스가 하나 그려지고(이 부분이 50%)
# 정중앙값(중간값) 50부분이 녹샌선...
# 최소값 0, 최대값 100... 이 부분이 검은색 가로선이 그려짐

plt.figure(figsize=(7,6))
df.boxplot(column=['feature'])
plt.yticks(np.arange(0,101, 5))
plt.show()
```

```python
tips = sns.load_dataset('tips')
```

```python
# 1. total_bill 값을 가지고 tip 데이터의 중앙값과 50%안에 들어가는 tip값을 살펴보자.
plt.figure(figsize=(6,3))
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
tips.sort_values(by='total_bill', ascending=False)
```

```python
# 날짜별로 데이터를 분석
sns.boxplot(x=tips['day'],y=tips['total_bill'])
plt.show()
```

```python
sns.boxplot(x=tips['day'],y=tips['total_bill'],data=tips, hue='smoker')
plt.show()
```

```python
# plt.figure(figsize=(10,20))
tips.boxplot(column=['tip'], by=['day', 'sex'], rot=45)
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Lmplot

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker' ,data=tips)
```

# heatmap

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# pivot_table.. passengers
dd = flights.pivot_table(values='passengers', columns='year', index='month')
dd
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=dd)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(data=dd, annot=True, fmt='d')
# annot=True : 과학지수표기로 값을 띄어줌
# fmt='d' : annot를 정수로 띄어줌
plt.show()
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```

```python
X, y, w = make_regression(n_samples=50, n_features=1, bias=100, noise=10, coef=True, random_state=0)
```

```python
xx = np.linspace(-3, 3, 100)
y0 = w*xx + 100
plt.plot(xx, y0, 'r-')
plt.scatter(X, y, s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('make_regression 예제')
plt.show()
```

```python
# 내가짠거
def make_regression2(n_samples, bias, noise, random_state) :
    np.random.seed(random_state)
    
    X = np.array(np.random.rand(n_samples))
    w = np.array(15.89695836)#p.array(np.random.rand())
    y0 = w*X + bias
    e = np.random.normal(0, noise, n_samples)
    y = y0 + e
    return X, y, w
```

```python
X, y, w = make_regression2(n_samples=100, bias=100, noise=100, random_state=0)
```

```python
xx = np.linspace(-3, 3, 100)
y0 = w*xx + 100
plt.plot(xx, y0, 'r-')
plt.scatter(X, y, s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('make_regression2 예제')
plt.show()
```

```python
# 정답
def make_regression2(n_samples, bias, noise, random_state=1) :
    np.random.seed(random_state)
    X = np.random.normal(size=n_samples)
    w = 50* np.random.normal(size=1)[0]
    y0 = w*X + bias
    e = np.random.normal(scale=noise, size=n_samples)
    y = y0 + e
    return X, y, w
```

```python
X,y, w= make_regression2(50, 100, 10)
```

```python
xx = np.linspace(-3, 3, 100)
y0 = w*xx + 100
plt.plot(xx, y0, 'r-')
plt.scatter(X, y, s=100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('make_regression2 예제')
plt.show()
```

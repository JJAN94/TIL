# AND 게이트 

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-8, 8, 100)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def ReLU(x):
    return np.where(x>0, x, 0)
```

```python
fig, ax = plt.subplots(nrows= 1, ncols=3, figsize=(20,5))
y_ls = [sigmoid(x), tanh(x), ReLU(x)]
y_title = ['sigmoid', 'tanh', 'ReLU']
for i in range(3):
    ax[i].plot(x, y_ls[i])
    ax[i].set_title(y_title[i])
```

```python
def Odds(x):
    return x/(1-x)

def Logit(x):
    return np.log(x/(1-x))

def Sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
fig, ax = plt.subplots(nrows= 1, ncols=3, figsize=(20,5))
y_ls = [Odds(x), Logit(x), Sigmoid(x)]
y_title = ['Odds', 'Logit', 'sigmoid']
for i in range(3):
    ax[i].plot(x, y_ls[i])
    ax[i].set_title(y_title[i])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(100)
data = {'a':np.arange(50),
        'c':np.random.randint(0,50,50),
        'd':np.random.randn(50)
       }
# data
```

```python
data['b'] = data['a']+10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100
```

# DataFrame 시각화

```python
df = pd.DataFrame(data)
df.head()
```

# 산점도 그래프
- scatter 함수를 사용한다.
- x, y축에 해당하는 데이터의 상관관계... 얼마나 많은 분포 혹은 흩어져 있는지를 한눈에 확인 가능
- x, y 두개의 축을 기준으로 데이터가 얼마나 퍼져있는지를 알 수 있다.
- legend 사용 불가

```python
# s: marker size
# c : marker color
# cmap : 'jet'
plt.scatter([1,2,3,4], [10,30,20,40], s=[100,200,350,500], c=range(4), cmap='hsv')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter('a', 'b', data=df, c='c', s='d')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter('c', 'd', data=df, c='c', s='d', cmap='hsv')
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter('b', 'd', data=df, c='c', s='d', cmap='spring')
plt.colorbar()
plt.show()
```

```python
cutoff=(df['a']>25) & (df['b']>20)

df['color'] = np.where(cutoff==True, 'r', 'b')
plt.figure(figsize=(12,6))
# sns.set_style('white')
sns.regplot(x=df['a'], y=df['b'], scatter_kws={'facecolors':df['color']})


plt.title('Scatter and Plot Regression', fontsize=16)
plt.show()
```

# subplot으로 plot(), scatter(), bar() 동시에 그려넣기
Subplot은 영역을 쪼개어서 여러개의 그래프를 표현하고 싶을 때 사용

```python
names = ['group-a', 'group-b', 'group-c'] # x값으로 이용
values = [1, 10, 100] # y값으로 이용
```

```python
plt.figure(figsize=(9,3))
plt.subplot(1,3,1)
plt.plot(names, values)

plt.subplot(1,3,2)
plt.bar(names, values)

plt.subplot(1,3,3)
plt.scatter(names, values)
plt.show()
```

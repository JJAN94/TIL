# 1. 스택/큐

## -기본개념-

#### 스택(Stack)
- 스택 자료구조는 책을 쌓는 것처럼 차곡차곡 쌓아 올린 형태의 자료구조를 말합니다.
- 스택에서 삽입하는 연산을 `push`, 삭제하는 연산을 `pop`이라고 합니다.
- 스택은 시간 순서에 따라 자료가 쌓여서 가장 마지막에 삽입된 자료가 가장 먼저 삭제되는 구조적인 특징을 가지고 있습니다.

#### 큐(Queue)
- 큐는 놀이동산에서 줄을 서서 기다리는 것처럼 먼저들어온 자료가 먼저 나가는 자료구조를 말합니다.
- 큐는 한쪽에서 삽입과, 삭제가 이뤄지는 스택과 달리 삭제연산만 수행되는 곳과, 삽입연산만 수행되는 곳으로 나뉘어져 있습니다.

## -특강-

#### 스택이란?
- 영어로 stack '쌓다'라는 의미
- 프로그래밍에서 목록 혹은 리스트에 접근이 한 쪽에서만 가능한 구조
- LIFO (Last In, First Out)가 기본원리
- push, peek, pop 함수

#### 스택의 구조

- BOOKS = [BOOK_1, BOOK_2, BOOK_3]이라는 리스트가 있다. 아래서부터 BOOK_1, BOOK_2, BOOK_3가 차곡차곡 쌓여있다.
    - BOOK_4를 추가하고 싶다 -> push(python에서 list.append와 기능 동일)
    - 맨위의 것을 확인하고 싶다 -> peek
    - 맨위의 것을 빼고 싶다 -> pop

#### Python 스택의 구현방법

1. 직접구현
2. 이미 구현된 클래스 import
3. List를 스택으로 구현
---
수업에서는 직접구현만 다룬다. 왜냐하면 Python에서는 List가 스택으로 사용 가능하도록 잘 구현되어있어서.

#### Python 스택 직접 구현

```python
class Stack(list) :
    push = list.append
    def peek(self) :
        return self[-1] # 또는 self[len(self)-1]
#     def pop(self) :
# pop함수는 만들지않는다. Python에 list 내장함수로 이미 구현되어있다.
```

```python
s = Stack() # Stack 객체선언
s.push(1)
s.push(5)
s.push(10)
print('My Stack is :', s)
print('popped value is :', s.pop())
print('My Stack is :', s)
print('peeked value is :', s.peek())
print('My Stack is :', s)
```

#### Python List를 스택으로 활용

```python
s = []
s.append(1)
s.append(5)
s.append(10)
print("My Stack is : ",s)
print('popped value is : ', s.pop())
print("My Stack is : ",s)
print('peeked value is : ', s[-1])
print("My Stack is : ",s)
```

#### 스택의 활용

- 브라우저의 앞, 뒤 페이지로 이동하기/ 이전페이지, 다음페이지 스택이 존재함.
- 깊이 우선 탐색(DFS)

#### 큐란?
- 영어로 Queue '일이 처리되기를 기다리는 리스트'라는 의미
- 프로그래밍에서 목록 혹은 리스트에서 접근이 양쪽에서 가능한 구조로 FIFO(First In, First Out)가 기본원리
- 대표함수 : put, peek, get

#### 큐의 구조
- 컨베이어 벨트위에 순서대로 Box1, Box2, Box3를 놓는다.
    - Box_ls = [Box1, Box2, Box3]
- 이제 추가로 Box4를 놓는다. -> put
    - Box_ls = [Box1, Box2, Box3, Box4]
- 가장 먼저 놓인 박스를 확인하고 싶다. -> peek
- 이제 컨베이어를 따라 박스를 트럭에 싣자. -> get
    - 맨처음 박스를 싣자. Box_ls = [Box2, Box3, Box3]

#### Python 큐의 구현방법
- 직접구현
- 이미 구현된 클래스 import
- List를 큐로 구현

#### Python 큐 직접 구현

```python
class Queue(list) :
    put = list.append
    def peek(self) :
        return self[0]
    def get(self) :
        return self.pop(0) # pop()에 parameter=0을 주면 idx=0을 추출
```

```python
q = Queue()
q.put(1)
q.put(5)
q.put(10)
print('My Queue is : ', q)
print('removed value is : ', q.get())
print('My Queue is : ', q)
print('peeked value is : ', q.peek())
print('My Queue is : ', q)
```

#### Python 구현된 클래스 import

```python
from queue import Queue
q = Queue()
q.put(1)
q.put(5)
q.put(10)
print("My Queue is : ", q)
print("removed value is :", q.get())
print("My Queue is : ", q)
print("peeked value is : ", q.peek) # ??? 강의에선 쓰는데 Queue class는 peek이 없다는데요??
print("My Queue is : ", q)
```

#### Python List를 큐로 활용

```python
q= []
q.append(1)
q.append(5)
q.append(10)
print("My Queue is : ",q)
print('removed value is : ', q.pop(0))
print("My Queue is : ",q)
print('peeked value is : ', q[0])
print("My Queue is : ",q)
```

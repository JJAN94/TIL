# web crawling
- 1. requests : json : 동적페이지(url의 변경 없이 데이터를 수정 및 추가)
- 2. requests : html : 정적페이지(url 변경으로 데이터를 수정 및 추가)
- 3. selenium : web browser : 1, 2번 방법을 사용하지 못할 때 사용
    - requests를 통해 요청시 서버에 요청이 많아져(파이썬으로) 차단 가능성 있음.
    - selenium으로 하면 웬만해선 다 크롤링이 가능하나 단점이 속도가 느리다.
    - 1->2->3번순으로 해보자. (1이 안되면 2번, 2가 안되면 3번...)
---
- 웹크롤링을 하고자 할 때, 혹 그 사이트에 모바일 페이지가 있는지 확인해보자. 모바일 페이지가 pc페이지보다 상대적으로 간단하기에 편하다

    2. requests, response : json 데이터(문자열)를 얻기

```python
import requests
import pandas as pd
```

```python
# https://m.stock.naver.com/sise/siseIndex.nhn?code=KOSPI
# 개발자도구(F12) -> Newtork -> XHR
# https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code=KOSPI&pageSize=20&page=6
```

```python
# 1. 웹 서비스 분석 : URL 찾기
code, page_size, page = "KOSPI", 20, 1

url = 'https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code={}&pageSize={}&page={}'\
.format(code, page_size, page)

print(url)
```

```python
# 2. requests, response : JSON(str)
response = requests.get(url)
response
```

```python
type(response.text) # response(json)의 타입
```

```python
# 3. JSON(str) -> JSON(dict) -> DF
# response.text
# 텍스트 데이터를 딕셔너리로 파싱해주는 함수가 있다.
datas = response.json()['result']['siseList']
kospi_df = pd.DataFrame(datas)
kospi_df.tail(3)
```

```python
# 함수로 만들기
def get_stock_data(code, page_size=20, page=1) :
    url = 'https://m.stock.naver.com/api/json/sise/dailySiseIndexListJson.nhn?code={}&pageSize={}&page={}'\
.format(code, page_size, page)
    response = requests.get(url)
    datas = response.json()['result']['siseList']
    return pd.DataFrame(datas)
```

```python
kosdaq_df = get_stock_data('KOSDAQ',100)
kosdaq_df.tail(3)
```

```python
kospi_df = get_stock_data('KOSPI', 100)
kospi_df.tail(3)
```

```python
# 3. JSON(str) -> DF
datas = response.json()
usd_df = pd.DataFrame(datas)
usd_df.tail(3)
```

```python
print(usd_df["localTradedAt"].values)
```

```python
usd_df.rename(columns={"localTradedAt":"dt"}, inplace=True)
usd_df.tail(3)
```

```python
dtt = []
dttt = []
for i in range(100) :
    dtt.append(usd_df['dt'].values[i].split('-'))
for i in range(100) :
    dttt.append(dtt[i][0] + dtt[i][1] + dtt[i][2])
```

```python
usd_df['dt'] = dttt
usd_df.tail(3)
```

```python
usd_df.rename(columns={'closePrice' : 'nv'}, inplace=True)
```

```python
print(kospi_df['dt'].values)
```

```python
# 데이터 합치기
merge_df_1 = pd.merge(kospi_df, kosdaq_df, left_on="dt", right_on='dt')
merge_df_1 = merge_df_1.rename(columns={"ncv_x" : "ncv_kospi", "ncv_y" : "ncv_kosdaq"})
merge_df_1 = merge_df_1[["dt", "ncv_kospi", "ncv_kosdaq"]]
merge_df_1.tail(2)
```

```python
merge_df = pd.merge(merge_df_1, usd_df)[['dt', 'ncv_kospi', 'ncv_kosdaq','nv']]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,5))

plt.plot(merge_df['dt'], merge_df['ncv_kospi'], 'r*:',label='kospi')
plt.plot(merge_df['dt'], merge_df['ncv_kosdaq'], 'bo--',label='kosdap')
plt.plot(merge_df['dt'], merge_df['nv'], 'g^-',label='usd')

# xtick이 항상 10개 출력되도록
step = len(merge_df) // 10
plt.xticks(merge_df[::step]['dt'])


plt.show()
```

```python
import numpy as np
```

```python
a = []
b = []
for i in range(100) :
    a.append(merge_df['nv'][i].split(','))
for j in range(100) :
    b.append(float(a[j][0]+a[j][1]))
merge_df['nv'] = b    
```

```python
merge_df.head(5)
```

```python
# 정규화 : z = (x- min(x)) / (max(x) - min(x))
kospi_max = np.max(merge_df['ncv_kospi'])
kospi_min = np.min(merge_df['ncv_kospi'])
kosdaq_max = np.max(merge_df['ncv_kosdaq'])
kosdaq_min = np.min(merge_df['ncv_kosdaq'])
usd_max = np.max(merge_df['nv'])
usd_min = np.min(merge_df['nv'])

normal_kospi = (merge_df['ncv_kospi']-kospi_min)/(kospi_max - kospi_min)
normal_kosdaq = (merge_df['ncv_kosdaq']-kosdaq_min)/(kosdaq_max - kosdaq_min)
normal_usd = (merge_df['nv']-usd_min)/(usd_max - usd_min)
```

```python
plt.figure(figsize=(20,5))

plt.plot(merge_df['dt'], normal_kospi, 'r*:',label='kospi')
plt.plot(merge_df['dt'], normal_kosdaq, 'bo--',label='kosdap')
plt.plot(merge_df['dt'], normal_usd, 'g^-',label='usd')

# xtick이 항상 10개 출력되도록
step = len(merge_df) // 10
plt.xticks(merge_df[::step]['dt'])


plt.show()
```
